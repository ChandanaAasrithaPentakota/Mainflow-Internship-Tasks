```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('householdtask3.csv')
df.head()
```

```python
plt.bar(df['year'],df['own'])
plt.title("BAR CHART")
plt.xlabel("YEAR")
plt.ylabel("OWN")
plt.show()
```

```python
plt.plot(df['year'])
plt.plot(df['own'])
plt.title("LINE CHART")
plt.xlabel("YEAR")
plt.ylabel("OWN")
plt.show()
```

```python
plt.scatter(df['year'],df['own'])
plt.title("SCATTER PLOT")
plt.xlabel("YEAR")
plt.ylabel("OWN")
plt.show()
```

```python
plt.hist(df['income'])
plt.title("HISTOGRAM")
plt.xlabel("YEAR")
plt.ylabel("OWN")
plt.show()
```
